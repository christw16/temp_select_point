# 大致上ｘ一排最多可以放三個 （包含overlap) ; 最少兩個
# 下面輸出是座標（x, y)

```python
# roi 大小 這裡是25*25
unit = 25

# 可以overlap的長度
remain = 10

# 輸入座標
x_start = 1
x_end = 79
y_start = 121
y_end = 206


def calculate(x_start, x_end, y):
    first_pair = []  
    second_pair = []
    third_pair = []
    
    #            示意圖
    #              78
    #    =====================
    #    ||     ||     ||    ||
    # 25 ||     ||     ||    ||
    #    ||     ||     ||    ||
    #    ======================
    #       25            
    
    # 我們討論的最左的正方形
    first_pair.append(x_start)
    first_pair.append(y)
    
    # 我們討論的最右的正方形
    second_pair.append(x_end-unit)
    second_pair.append(y)

    # 我們討論的中間的正方形. 上面那個是示意圖，夾在中間的正方形一定會overlapped
    third_x1 = x_end -(unit-remain)
    third_x2 = first_pair[0] + (unit-remain)
    difference = third_x1 - third_x2

    if ((difference / unit) >= 1 and (difference / unit) < 2):
        third_start = unit - remain
        third_end = x_end - (unit-remain) - unit
        third_pair.append(random.randint(third_start, third_end))
        third_pair.append(y)
        print("最左邊正方形")
        print(first_pair)
        print("中間可能會overlapped正方形")
        print(third_pair)
        print("最右邊正方形")
        print(second_pair)
        print("滿了 換下一行")
        print('\n')
    else:
        print(first_pair)
        print(second_pair)
        print("there is no middle one, it can only select two square")
        print("================================================")
    
    return first_pair[1]
        
    
for i in range(100):
    print("round: ",i+1)
    a = calculate(x_start, x_end, y_start)
    y_start += 10
    y_start = y_start + remain
    
    if (a >= y_end):
        break


```
